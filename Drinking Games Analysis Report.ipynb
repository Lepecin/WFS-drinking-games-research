### 2.1  Twenty-One
'Twenty-One', also known as 'Twenty-One Dares', involves a group of people arranged in a circle taking turns to collectively count up to a value, which is 21 in this case.
The person who says "21" has to drink in the drinking game version of 'Twenty-One'. This means game has two possible objectives, either you aim to say "21" or you avoid saying "21".
The process of counting up to 21 depends on the rule set. This report will only consider rules that affect the counting process and the order of the turns the players take.

### 2.2 Spoons
'Spoons' involves people arranged in a circle, each with a set number of randomly dealt playing cards. Each takes a card discarded by the person on the right (or left) to replace one of their own cards with it. The replaced card is discarded and passed to the person on the left (or right).
The aim is to collect the maximal amount of cards of the same value.
This report will use a computer simulation of 'Spoons' to analyse it, as it has a mathematically chaotic structure.

## 3. Twenty-One Game Analysis

### 3.1 Mathematical Model
The game requires $n$ people aranged in a circle, where $n \in \mathbb{N}$.
Each player is numbered $0$ to $n-1$. The first player who counts is chosen to be player $0$.

The game is modelled to be a sequence of states. The game's state at period $t$ represented by a vector $v_t \in (\mathbb{Z}/n\mathbb{Z})\times \mathbb{N}_0$, where $\mathbb{Z}/n\mathbb{Z}$ is the ring of integers modulo $n$.
The first entry of $v_t$ denotes number of the player counting at period $t$, and the second entry denote the number collectively counted up to in all periods before period $t$.
At $t=0$, the game state is $v_0=(0,0)$, since the game starts with player $0$ counting and the count starts at $0$.

Here, we generalise the counting rules as follows.
At each state, the counting player can choose how many numbers to count by. This choice is limited by their decision space $\left \{ d_1,d_2, \ldots, d_k \right \}=D \subseteq \mathbb{N}$ where each $d_i$ is an amount by which they can count.
We introduce the rule function $f:D \mapsto \mathbb{Z}/n\mathbb{Z}$, which determines how each decision from $D$ changes the player who will count.

Suppose the game state at period $t$ is $v_t=(p_1,p_2)$, so player $p_1$ is counting on top of the number $p_2$. If the player chooses to count by $x_t \in D$, where $x_t$ is the decision made at period $t$, then the game state at period $t+1$ will be $v_{t+1}=(p_1+f(x_t),p_2+x_t)$.
In general:<br>
<br>
<center>
$v_{t+1}=\left (\sum_{i=0}^tf(x_i), \sum_{i=0}^tx_i \right ):\forall t \in \mathbb{N}_0$
</center>

### 3.2 Specific Versions
There are two popular versions of 'Twenty-One' which are described as follows.

**Version 1**: One step 'Twenty-One'
Each player can count up by either one, two or three numbers. Regardless of how many numbers a player counts, the player coming directly after them is the next player who counts. Thus, the decision space for each player is $D = \left \{ 1,2,3 \right \}$ and the rule function $f$ outputs $1$ for any input element from $D$.

**Version 2**: Drinking game 'Twenty-One'
Each player can count up by either one, two, three or four. Counting up by one passes the turn to the next player, while counting up by two passes the turn to the previous player. Counting up by three or four would be the same as counting up by one or two respectively, with the added twist of skipping a player. Therefore, the decision space is $D = \left \{ 1,2,3,4 \right \}$ and the rule function $f$ has the following summary:<br>
<br>
<center>
$f(x)=\begin{cases}
1 & \text{ if } x=1 \\ 
-1 & \text{ if } x=2 \\ 
2 & \text{ if } x= 3\\ 
-2 & \text{ if } x=4 
\end{cases}$
</center>

For both versions, the game is lost (or won) by the player who counts up to 21 inclusive.

### 3.3 Game Theoretic Analysis Example
From here, we will assume that each player is avoiding to count up to some number N. This means all players are aiming to count up to $N-1$, though it's not necessary to avoid counting to $N$ as analysis will demonstrate.

By rules of the first version of 'Twenty-One', the main player (you) can count up by one, two or three, while your opponents can collectively count up by an amount between $n-1$ and $3(n-1)$ before it's your turn again.

We now look for the conditions where the main player can guarantee a win. Using backwards induction, it's obvious that the main player loses if they start counting up after $N-1$; they are bound to count $N$. 
However, avoiding a loss can be guaranteed if the main player counts up after any value between $N-2$ and $N-4$ for at least two players, as they can count up to $N-1$, forcing the next player to count $N$ and lose.

Further to this, the player can guarantee avoiding a loss if they count up after any value $N-2$ and $N-2-n$ when playing an $n$ player game. A minimum of $n-1$ additional counts must be collectively done by your $n-1$ opponents before it's your turn again, so by counting up by 3 after $N-2-n$, there are $n-1$ counts left until someone counts $N$, thus one of your opponents is bound to lose.<br>
<br>
<center>
$\overbrace{N-1}^{\text{Loss}},\underbrace{N-2,N-3,\ldots,N-2-n}_{\text{Loss avoidable}},\overbrace{N-3-n,\ldots}^{\text{Can lose}}$
</center>

If the number of players is more than two, there are no other numbers to count up after where a loss is avoidable except the ones mentioned above. The two player case is special, in that there are more values where you can avoid a loss. In fact, loss is avoidable when counting up after any interger in the range $N-2-4k$ and $N-4-4k$ inclusive for $k \in \mathbb{N}_0$. You can check it yourself.<br>
<br>
<center>
$\overbrace{N-1}^{\text{Loss}},\underbrace{N-2,N-3,N-4}_{\text{Loss avoidable}},\overbrace{N-5}^{\text{Can lose}},\underbrace{N-6,N-7,N-8}_{\text{Loss avoidable}},\ldots$
</center>

Setting $N=21$, we see that, for two players, loss is always avoidable when counting up after any number underlined in the sequence below:<br>
<br>
<center>
$20,\underline{19,18,17},16,\underline{15,14,13},12,\underline{11,10,9},8,\underline{7,6,5},4,\underline{3,2,1},0$
</center>

Evidently, you can always avoid a loss by playing second.

### 3.4 Generating Function Analysis Example
The rules of the second version of 'Twenty-One' are trickier to work with, but we can try asking a different question now. What if we want to find the number of ways of in which a game can start with one person and end with another given the number up to which the players counted?

Here, we demonstrate the use of a matrix to create generating functions that yeild us such combinatorial answers. Let matrix $A$ have polynomial elements, where $A_{ij}=P_{ij}(x)$ is a polynomial (a generating function) for the number of ways in which a single turn starts with player $i$ and ends with player $j$, where the exponents of the polynomial represent the final number up to which the start player counted.

For simplicity, we will look at the drinking game version with three players. 

There is no way for a player to end up back at themselves in one turn. So, for every player, $A_{ii}=P_{ii}(x)=0: \forall i \in \mathbb{Z}/3\mathbb{Z}$.

For a player to end with the next player in one turn, they can either count up by one or by four. Since there is only one way of counting up by either one or four, the following equations hold:<br>
<br>
<center>
$A_{01}=P_{01}(x)=x+x^4=A_{12}=A_{20}$
</center>

Lastly, a player can end with the previous player in one turn if they either count up by two or three. Again, there is only one way of counting up by three or two, therefore:<br>
<br>
<center>
$A_{02}=P_{02}(x)=x^2+x^3=A_{10}=A_{21}$
</center>

The final generating function matrix $A$ has the form:<br>
<br>
<center>
$A=\begin{pmatrix}
A_{00} & A_{01} & A_{02}\\ 
A_{10} & A_{11} & A_{12}\\ 
A_{20} & A_{21} & A_{22}
\end{pmatrix}
=
\begin{pmatrix}
0 & x+x^4 & x^2+x^3\\ 
x^2+x^3 & 0 & x+x^4\\ 
x+x^4 & x^2+x^3 & 0
\end{pmatrix}$
</center>

We can now, for example, analyse the matrix $A^3$ to find the number of ways the game can start and end with the second player in three turns, given that the final count is $6$. For this, we look at the generating function formed at $(A^3)_{22}$, which is $P^3_{22}= x^3+4x^6+3x^7+3x^8+4x^9+x^12$. Since the final count is $6$, the coefficient for the $x^6$ term is $4$, so there are $4$ ways of starting and ending with player two in three turns and having the final count equal $6$.

This is a powerful combinatorial tool that also has applications in graph theory. Indeed, 'Twenty-One' can be expressed as a graph, where the nodes are the players connected with directed veritices, and each vertex weighted by the count number required for the game to swich from one player to another.

## 4. Spoons Game Analysis

### 4.1 Mathematical Model
Considering a game of 'Spoons' with:
- $t$ players
- $m$ suits of cards
- $n$ different values the cards can take
- $k$ being the number of cards a player must hold at any time

From this, the inequality governing the values is $tk \leq nm$, since the players can't collectivly hold more cards than there are.

Define the vector $p_i=(p_{i1},p_{i2},\ldots,p_{in}) \in \mathbb{N}_0^n$ to represent the number of different card values that player $i$ is holding. So $p_{ij}$ equals the number of cards of value $j$ that player $i$ is holding. Each player has to hold $k$ cards at any time, so the following constraint holds:<br>
<br>
<center>
$\sum^n_{j=1}p_{ij}=\mathbf{1} \cdot p_i=k:\forall i \in \left \{ 1,2,\ldots,t \right \}$
</center>

Where $\mathbf{1}$ is the unit vector, with all entries equal to one.
The aim of the game, for any player $i$ is to have only one entry in $p_i$ equal $k$, which creates another restriction $k \leq m$.

Finlly, introduce the vector $p_0=m\mathbf{1}-\sum^{t}_{i=1}p_i$, which represents the number of cards of different values in the main deck.

### 4.2 The Optimal Strategy
A single player can treat the entire situation as though they are fed random cards from a deck of $nm-k$ cards, since they can recieve cards from either the main deck or the other $t-1$ players. This is, of course, a simplification, but the game would be significantly harder to model by accounting the possible dsitributions of cards among all players.

It may be obvious to prioritse keeping the cards that increase the greatest number cards held of the same value and discard cards that would decrease the smallest number of cards held of the same value, but is there a way show that it's the best? 

For player $i$, under the previous assumptions, the probability of getting a card of value $j$ is $(m-p_{ij})/(nm-k)$ if they are already holding $p_{ij}$ cards of that value. With this known, the probability of getting $k-p_{ij}$ cards of value $j$ in the next $k-p_{ij}$ turns is:<br>
<br>
<center>
$\mathbb{P}=\prod ^{k-p_{ij}}_{r=1}\dfrac{m-p_{ij}+r-1}{nm-k}=\dfrac{(k-p_{ij})!}{(nm-k)^{k-p_{ij}}}\binom{m-k}{m-p_{ij}}$
</center>

Since $\mathbb{P}$ gets smaller as $p_{ij}$ reduces, it is better to keep cards that increase the greatest value in $p_{i}$ and discard cards that reduce the smallest value in $p_{i}$.

### 4.3 Small Modifications
The optimal strategy dicussed previously still has nuances to resovle.

In the case of an even $k$ and $k=m$, it's possible that a player can have half of their cards being of one value and the other half of another value. The problem occurs when a second player attains the same combination of cards, meaning both players will never win if they play the optimal strategy. Despite it being very unlikely, it can make the game endless if the number of players $t$ is even.

Another detail to consider is the bias towards discarding certain cards according to their value. If players are going to prefer holding higher value cards over lower value ones, the lack of diversity of card values in circulation could lead to longer games. A player could be holding back certain cards that are useful only to another player, thus reducing the actual probability that their opponents could get certain cards, meaning they could as easily fall victim to this issue as well.

Both issues are resolved by occasional randomisations in the players' behaviours with regards to what cards they discard. If a player has an equal amount of two different valued cards, they will replace one of their cards with the one passed down to them. If a player has multiple categories of card values that have the smallest number of cards, the player will randomly choose the category to discard from.

### 4.4 Analysis Through Simulation
By running 100000 simulations of 'Spoons' using R code and the mathematcial model described above with the modified optimal strategy, it becomes evident that each player does not have the same chance of achieving $k$ indentically valued cards before everyone else.
